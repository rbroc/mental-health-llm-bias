```python
import pandas as pd
```

```python
path_gpt_completions = '/home/Plaza/mental-health-llm-bias/evaluation/data/model_completions/gpt-4/phq-9_final.csv'
data = pd.read_csv(path_gpt_completions)
```

```python
data['response_condition'].value_counts()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)  # To ensure long text is fully visible

# Group by 'response_condition' and count 'model_completion' values
model_completion_counts = data.groupby('response_condition')['model_completion'].value_counts().reset_index(name='count')

# Format the output with tabs between columns
formatted_output = model_completion_counts.apply(lambda row: f"{row['response_condition']}\t{row['model_completion']}\t{row['count']}", axis=1)

# Print each formatted line
for line in formatted_output:
    print(line)
```

```python
# Check for model refusals
```

```python
pattern = r'(as an AI|I don\'t have feelings|only a qualified healthcare professional|consult with a healthcare professional|reach out to a mental health professional|seek help immediately)'
```

```python
# Check for refusals
data['is_refusal'] = data['model_completion'].str.contains(pattern, case=False, regex=True)

# Count the refusals
refusal_count = data['is_refusal'].sum()
print("Number of model refusals:", refusal_count)
```
